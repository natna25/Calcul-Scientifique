## résolution de $\nabla f(x) =0$

Dans cette partie, on considère le problème de minimisation:
$$ min_{x\in \mathbb{R}^n}(f(x))$$

On sait que la solution existe lorsqu'on vérifie 
$$ \nabla f(x)=0$$

```python
import numpy as np
import matplotlib.pyplot as plt

```

on clacule les itérations des valeurs à partir de la formule:

$$ X^{n+1} = X^n - [H_f(X^n)]^{-1} \nabla f(X^n) $$

où $H_f(X^n)$ est la matrice hessienne de f évalué au point $X^n$

on remarque également que cette méthode n'est pas définie si la hessienne n'est pas inversible

aussi, calculer $ [H_f(X^n)]^{-1} \nabla f(X^n)$ revien à calculer $ H_f(X)d= - \nabla f(X)$

## Minimisation de la fonction de Rosenbrock

la formule de rosenbrock est une fonction non-convexe qui est souvent utilisé dans les problèmes d'optimisation car elle contient plusieurs minimum locaux. Il convient alors de trouver son minimum global au point (1,1)

```python
def rosenbrock(x,y):
    return (np.power(1-x,2) + 100*np.power(y-np.power(x,2),2) )
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm


step = 0.1
x_grid,y_grid = np.mgrid[-2:2:step, -1:3:step]

fig = plt.figure()
ax = fig.gca(projection='3d')

#axis labels
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$rosenbrock$')
surf = ax.plot_surface(x_grid, y_grid, rosenbrock(x_grid,y_grid), cmap=cm.viridis,
                   linewidth=0, antialiased=True)
plt.show()
```

```python
def hess_rosenbrock(x,y):
    
    line1 = [1200*np.power(x,2) - 400*y + 2 , -400*x]
    line2 = [-400*x , 200]
    
    return np.array([line1,line2])
```

```python
def grad_rosenbrock(x,y):
    
    df_dx = 2*x - 2 - 400 * (x*y - np.power(x,3) )
    
    df_dy = 200*(y-np.power(x,2))
    
    return np.array([df_dx,df_dy])
```
