On chercher à calculer le prix d'un call européen avec comme valeurs :

- Maturité : T = 1

- Strike : K= E = 100

- Taux constant (celui de la banque centrale européenne) : r = 0.1

- La volatilité : $\sigma = 0.3 $ 

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from scipy.special import erfc
```

```python
def Black_Scholes(S,E,r,t,sigma):
    
    d1 = (np.log((S/E)) + (r + (sigma**2)/2) * T)/ (sigma * sqrt(T) )
    #print(d1)
    d2 = d1 -( sigma * sqrt(T))
    n1 = 0.5 * erfc(-d1/sqrt(2))
    n2 = 0.5 * erfc(-d2/sqrt(2))
    
    call = ((S*n1) - (E * np.exp(-r*T) * n2 ) )
    #print(call)
    
    
    return call
```

```python
sigma = 0.3
n= 200
E=100
r = 0.1
T = 1
L = np.log(3)
h = 2/(n-1) #step

x = np.arange(-1,1,h)

S = E * np.exp(L*x)



call = []
    
call.extend(Black_Scholes(S,E,r,T,sigma))
```

```python
plt.plot(S,call)
```

sortie graphique 3d

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm


sigma = 0.3
N = 50
k = 100
r = 0.5
T= 1
e = 1e-4

Svals = np.linspace(e,3,N)
tvals = np.linspace(e,T-e, N)

[Smat,tmat] = np.meshgrid(Svals,tvals)

S = E * np.exp(L*Smat)

call = Black_Scholes(S,e,r,T,sigma)


fig = plt.figure()
ax = fig.gca(projection='3d')

#axis labels
ax.set_xlabel('$t$')
ax.set_ylabel('$s$')
ax.set_zlabel('$C(t,s)$')
surf = ax.plot_surface(tmat,Smat,  np.reshape(call, (Smat.shape)), cmap=cm.viridis,
                   linewidth=0, antialiased=True)
plt.show()


print(Svals.shape)
```
