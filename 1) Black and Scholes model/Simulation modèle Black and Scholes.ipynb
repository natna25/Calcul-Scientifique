On chercher à calculer le prix d'un call européen avec comme valeurs :

- Maturité : T = 1

- Strike : K= E = 100

- Taux constant (celui de la banque centrale européenne) : r = 0.1

- La volatilité : $\sigma = 0.3 $ 

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from scipy.special import erfc
```

```python
def Black_Scholes(S,E,r,t,sigma):
    
    d1 = (np.log(S/E) + (r + (sigma**2)/2) * T)/ (sigma * sqrt(T) )
    print(d1)
    d2 = d1 - sigma * sqrt(T)
    n1 = 0.5 * erfc(-d1/sqrt(r))
    n2 = 0.5 * erfc(-d2/sqrt(r))
    
    call = ((S*n1) - (E * np.exp(-r*T) * n2 ) )
    #print(call)
    
    
    return call
```

```python
sigma = 0.3
n= 200
E=100
r = 0.1
T = 1
L = np.log(3)
h = 2/(n-1) #step

x = np.arange(-1,1,h)

S = E * np.exp(L+x)

call = []
for s in S:
    
    call.extend(Black_Scholes(S,E,r,T,sigma))
    break
    
print(call)

```
